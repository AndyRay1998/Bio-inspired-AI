```python
import numpy as np
import matplotlib.pyplot as plt
import random
import operator
%matplotlib inline
```

# initialization

```python
# size of space
n_x = 100
n_y = 100
space = np.zeros((n_x, n_y), dtype=np.float)
# seed point
space[int(n_x/2)-1][int(n_y/2)-1] = 255
```

```python
for i in range(1000):
    rand_coor = random.sample(range(min(n_x, n_y)), 2)
    x_temp = rand_coor[0]
    y_temp = rand_coor[1]
    
    print(i)
    
    while(np.sum(space[x_temp-1:x_temp+2, y_temp-1:y_temp+2])==0):
        
        x_temp = np.clip(x_temp, 1, n_x-3)
        y_temp = np.clip(y_temp, 1, n_y-3)

        direction = random.sample(range(1, 9), 1)[0]
        # print(np.sum(space[x_temp-1:x_temp+1][y_temp-1:y_temp+1]))
        # print(f'{x_temp}, {y_temp}')
        if(direction == 1):
            x_temp += 1
        elif(direction == 2):
            x_temp += 1
            y_temp += 1
        elif(direction == 3):
            y_temp += 1
        elif(direction == 4):
            x_temp -= 1
            y_temp += 1
        elif(direction == 5):
            x_temp -= 1
        elif(direction == 6):
            x_temp -= 1
            y_temp -= 1
        elif(direction == 7):
            y_temp -= 1
        elif(direction == 8):
            x_temp += 1
            y_temp -= 1
    
    space[x_temp][y_temp] = 255
```
