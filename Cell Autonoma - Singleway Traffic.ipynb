```python
import numpy as np
import matplotlib.pyplot as plt
import random
import copy
%matplotlib inline
```

# initialization

```python
length = 100
density = 0.2
traffic = np.zeros((1, length))
idx = random.sample(range(length), int(length*density))
for i in range(len(idx)):
    traffic[0][idx[i]] = 1
```

# state transform

```python
result = copy.deepcopy(traffic)

for j in range(100):
    print(j)
    
    traffic_next = np.zeros((1, length))
    
    for i in range(length):
        # deal with boundary
        if(i == 0):
            temp_tra = np.array([traffic[0][-1], traffic[0][0], traffic[0][1]])
        elif(i == length-1):
            temp_tra = np.array([traffic[0][-2], traffic[0][-1], traffic[0][0]])
        else:
            temp_tra = traffic[0, i-1:i+2]
        
        # state transform
        if(np.sum(temp_tra == np.array([1, 1, 1])) == 3):
            traffic_next[0][i] = 1
        elif(np.sum(temp_tra == np.array([1, 1, 0])) == 3):
            continue
        elif(np.sum(temp_tra == np.array([1, 0, 1])) == 3):
            traffic_next[0][i] = 1
        elif(np.sum(temp_tra == np.array([1, 0, 0])) == 3):
            traffic_next[0][i] = 1
        elif(np.sum(temp_tra == np.array([0, 1, 1])) == 3):
            traffic_next[0][i] = 1
            
    traffic = traffic_next
    
    # stack result for show
    result = np.vstack((result, traffic_next))
```
