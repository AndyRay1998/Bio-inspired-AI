```python
import numpy as np
import matplotlib.pyplot as plt
import random
import operator
import copy
%matplotlib inline
```

# initialization of population

```python
population = []
group_size = 1000
select_num = 600

# distance matrix
distance = np.array([[0, 10, 20, 30, 40],
                    [10, 0, 35, 17, 33],
                    [20, 35, 0, 9, 53],
                    [30, 17, 9, 0, 47],
                    [40, 33, 53, 47, 0]], dtype=np.float)

# random initialization
for j in range(group_size):
    
    popu_temp = random.sample(range(1, 5), 4)
    # start with 0
    popu_temp.insert(0, 0)
    
    # fitness calculation
    fit = 0
    for k in range(4):
        fit += distance[popu_temp[k]][popu_temp[k+1]]
    
    # put fitness to last in list
    popu_temp.append(fit)
    population.append(popu_temp)
```

# select and copy

```python
# choose most fit members
def snc():
    population.sort(key=operator.itemgetter(-1))
    for i in range(group_size-select_num):
        del(population[-1])
        # del(population[select_num])
        # temp_l = copy.deepcopy(population[i])
        # population.append(temp_l)
```

# gene cross

```python
def gc():
    for i in range(select_num, group_size):
        temp_l = [0]
        mut_index = random.sample(range(1, 5), 2)
        ind_index = random.sample(range(select_num), 2)
        
        temp_l.append(population[ind_index[0]][mut_index[0]])
        temp_l.append(population[ind_index[0]][mut_index[1]])
        
        for k in range(5):
            if(population[ind_index[1]][k] not in temp_l):
                temp_l.append(population[ind_index[1]][k])
        # a place for fitness
        temp_l.append(0)
        
        population.append(temp_l)
```

# mutation

```python
def m():
    for i in range(select_num, group_size):
        # mutation probability 0.01
        if(random.random() < 0.01):
            mut_index = random.sample(range(1, 5), 2)
            temp = population[i][mut_index[0]]
            population[i][mut_index[0]] = population[i][mut_index[1]]
            population[i][mut_index[1]] = temp
```

# update fitness

```python
# update fitness
def uf():
    for i in range(group_size):
        fit = 0
        for k in range(4):
            fit += distance[population[i][k]][population[i][k+1]]
        population[i][-1] = fit
```

# main

```python
if __name__ == "__main__":
    optimal = []
    for h in range(100):
        print(h)
        snc()
        gc()
        m()
        uf()
        optimal.append(population[0][-1])
```
