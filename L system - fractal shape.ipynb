```python
import numpy as np
import matplotlib.pyplot as plt
import copy
%matplotlib inline
```

```python
# F: 1  +: 2 -: 3
# F -> F+F--F+F
motion = [1]

x_list = [0]
y_list = [0]
phi = 0

for j in range(5):
    motion_temp = []
    for i in range(len(motion)):
        if(motion[i] == 1):
            motion_temp.append(1)
            motion_temp.append(2)
            motion_temp.append(1)
            motion_temp.append(3)
            motion_temp.append(3)
            motion_temp.append(1)
            motion_temp.append(2)
            motion_temp.append(1)
        else:
            motion_temp.append(motion[i])
            
    motion = copy.deepcopy(motion_temp)
```

```python
step = 1
for i in range(len(motion)):
    if(motion[i] == 1):
        x_list.append(x_list[-1] + step * np.cos(phi))
        y_list.append(y_list[-1] + step * np.sin(phi))
    elif(motion[i] == 2):
        phi += np.pi / 3
    elif(motion[i] == 3):
        phi -= np.pi / 3
```

```python
plt.plot(x_list, y_list)
```
