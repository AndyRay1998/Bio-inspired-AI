```python
import numpy as np
import matplotlib.pyplot as plt
import copy
%matplotlib inline
```

```python
# F: 1  +: 2 -: 3 [: 4 ]: 5
# F -> F[+F]F[-F[+F][-F]]F

motion = [1]

x_list = [0]
y_list = [0]
phi_list = [np.pi/2]


for j in range(4):
    motion_temp = []
    for i in range(len(motion)):
        if(motion[i] == 1):
            motion_temp.append(1)
            motion_temp.append(4)
            motion_temp.append(2)
            motion_temp.append(1)
            motion_temp.append(5)
            motion_temp.append(1)
            motion_temp.append(4)
            motion_temp.append(3)
            motion_temp.append(1)
            motion_temp.append(4)
            motion_temp.append(2)
            motion_temp.append(1)
            motion_temp.append(5)
            motion_temp.append(4)
            motion_temp.append(3)
            motion_temp.append(1)
            motion_temp.append(5)
            motion_temp.append(5)
            motion_temp.append(1)
        else:
            motion_temp.append(motion[i])
            
    motion = copy.deepcopy(motion_temp)
```

```python
step = 1
stack = []
phi = np.pi/2

for i in range(len(motion)):
    if(motion[i] == 1):
        phi_list.append(phi)
        x_list.append(x_list[-1] + step * np.cos(phi_list[-1]))
        y_list.append(y_list[-1] + step * np.sin(phi_list[-1]))
    elif(motion[i] == 2):
        phi += 20 / 180 * np.pi
    elif(motion[i] == 3):
        phi -= 20 / 180 * np.pi
    elif(motion[i] == 4):
        stack.append(len(x_list)-1)
    elif(motion[i] == 5):
        x_list.append(x_list[stack[-1]])
        y_list.append(y_list[stack[-1]])
        phi_list.append(phi_list[stack[-1]])
        phi = phi_list[stack[-1]]
        del(stack[-1])
        
    print(f'{x_list[-1]}, {y_list[-1]}, {phi_list[-1]}')
```

```python
for i in range(int(len(x_list)/2)):
    plt.plot([x_list[i*2], x_list[i*2-1]], [y_list[i*2], y_list[i*2-1]])
plt.axis('equal')
```
