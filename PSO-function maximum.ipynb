Find maximal value of a single-parameter function

```python
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python
def obj_func(x):
    return - x ** 4 + 2 * x + 3
```

```python
def update_best(val, x_max_his, x):
    # update val
    val = obj_func(x)
    
    # global max
    maxindex  = np.argmax(val)
    max_x = x[maxindex]
    
    # local max
    for i in range(len(val)):
        if(obj_func(x_max_his[i]) < val[i]):
            x_max_his[i] = x[i]
            
    return max_x, x_max_his, val
```

# initialization

```python
num_par = 100
max_x = 50
min_x = -50
x = np.random.uniform(min_x, max_x, size=(num_par))
v = np.zeros(num_par)
x_max_his = np.zeros((num_par))
val = obj_func(x)
w = 0.8
c1 = 2
c2 = 2
max_x, x_max_his, val = update_best(val, x_max_his, x)
```

# main 

```python
if __name__ == '__main__':
    fittest = []
    for j in range(100):
        for i in range(len(val)):
            v[i] = w * v[i] + c1 * random.random() * (x_max_his[i] - x[i]) + c2 * random.random() * (max_x - x[i])
            x[i] += v[i]
        max_x, x_max_his, val = update_best(val, x_max_his, x)
        # print(x, v)
        # print(np.mean(x))
        fittest.append(obj_func(max_x))
```
